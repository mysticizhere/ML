```python
r = np.random.normal(0,1,1000)
```

```python
r
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(e,e, label='linear')
plt.plot(e,e**2,label='quadratic')
plt.plot(e,e**3,label='cubinc')
plt.legend()
plt.savefig('first.png')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(r, label='linear')
plt.legend()
plt.savefig('random.png')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
sr=np.sort(r)
plt.plot(sr, label='linear')
plt.legend()
plt.savefig('sorted.png')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(sr+0.2*r ,label='linear')
plt.legend()
plt.savefig('noise.png')
```
